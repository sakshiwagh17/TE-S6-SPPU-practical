```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('iris')
data.head()
```

```python
print("Data Types")
data.dtypes
```

```python
print('Features')
data.info()
```

```python
data.columns
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,9))
plt.suptitle("Histogram for iris dataset")
sns.histplot(data["sepal_length"],ax=axes[0,0])
sns.histplot(data["sepal_width"],ax=axes[0,1])
sns.histplot(data["petal_length"],ax=axes[1,0])
sns.histplot(data["petal_width"],ax=axes[1,1])
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,9))
plt.suptitle("Boxplot for each feature")
sns.boxplot(y="sepal_length" ,x="species",data=data,ax=axes[0,0])
sns.boxplot(y="sepal_width" ,x="species",data=data,ax=axes[0,1])
sns.boxplot(y="petal_length" ,x="species",data=data,ax=axes[1,0])
sns.boxplot(y="petal_width",x="species",data=data,ax=axes[1,1])
```
