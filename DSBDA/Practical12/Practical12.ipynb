```python
sample_document = "Tokenization is the process of breaking down text into words, phrases, symbols, or other meaningful elements. POS tagging assigns a part-of-speech tag to each word. Stop words removal removes common words that do not carry much meaning. Stemming and Lemmatization normalize words to their root form."
```

```python
nltk.download('all')
```

```python
# Tokenization
tokens = word_tokenize(sample_document)
tokens
```

```python
pos_tags=pos_tag(tokens)
pos_tags
```

```python
stop_words=set(stopwords.words('english'))
stop_words
```

```python
filtered_words=[word for word in tokens if word.lower() not in stop_words]
filtered_words
```

```python
stemmer =PorterStemmer()
stremmed_tokens=[stemmer.stem(word) for word in filtered_words]
stremmed_tokens
```

```python
lemmatization=WordNetLemmatizer()
lemmatization_words=[lemmatization.lemmatize(word) for word in filtered_words]
lemmatization_words
```

```python
tfidf_vectorize=TfidfVectorizer()
tfidf_matrix=tfidf_vectorize.fit_transform([' '.join(filtered_words)])
tfidf_matrix.to
```
