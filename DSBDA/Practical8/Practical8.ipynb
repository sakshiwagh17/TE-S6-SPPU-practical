```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import precision_score,r2_score,recall_score
```

```python
data=({"Experience":[5,7,9,12,18,20],"Salary":[20000,25000,40000,60000,80000,110000]})
df=pd.DataFrame(data)
df
```

```python
# Define independent (X) and dependent (y) variables
X=df[["Experience"]]
y=df["Salary"]
print(X)
print(y)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict salary for 15 years of experience
experience = np.array([[15]])
predicted_salary = model.predict(experience)

print(f"Predicted Salary for 15 years of experience: ₹{predicted_salary[0]:,.2f}")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(15, predicted_salary, color='green', label='Predicted Salary for 15 Years')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (₹)')
plt.title("Salary vs. Experience")
plt.legend()
plt.grid(True)
plt.show()
```
